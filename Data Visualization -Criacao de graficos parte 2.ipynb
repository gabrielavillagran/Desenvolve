```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# faz com que os graficos aparecam de forma automatica
%matplotlib inline

# Aumenta o tamanho das celulas
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))
```

```python
dados = pd.read_csv('./curso-matplotlib-preparando_ambiente/./iris.csv')
dados.head()
```

```python
# Criando a região do grafico
fig = plt.figure(figsize = (12,6))
ax = fig.add_axes([0, 0, 1, 1])

##### Grafico de dispersão #####
ax.scatter(dados['comprimento_sépala'], dados['largura_sépala'])

# Titulo e etiquetas
ax.set_xlabel('Comprimento da sépala',fontsize= 14)
ax.set_ylabel('largura da sépala',fontsize= 14)
ax.set_title('Comprimento da sépala x Largura Sépala', fontsize=20, pad = 15)
```

### Aumentando os ticks dos labels 

```python
# Criando a região do grafico
fig = plt.figure(figsize = (12,6))
ax = fig.add_axes([0, 0, 1, 1])

##### Grafico de dispersão #####
ax.scatter(dados['comprimento_sépala'], dados['largura_sépala'])

# Titulo e etiquetas
ax.set_xlabel('Comprimento da sépala',fontsize= 14)
ax.set_ylabel('largura da sépala',fontsize= 14)
ax.set_title('Comprimento da sépala x Largura Sépala', fontsize=20, pad = 15)

# Alterando o tamanho dos ticks dos labels
ax.tick_params(labelsize=15)
```

### Vamos ver quantas espécies diferentes temos e vamos diferenciá-las por cor 

```python
dados['espécie'].unique()

# temos 3 espécies diferentes
```

```python
# Criando a região do grafico
fig = plt.figure(figsize = (12,6))
ax = fig.add_axes([0, 0, 1, 1])

##### Grafico de dispersão #####

# Cores : diferenciando as especias por cores
cores = {'Iris-setosa':'darkorange', 'Iris-versicolor':'k','Iris-virginica':'b'}

for especie in dados['espécie'].unique():
    temporario = dados[dados['espécie'] == especie]
    ax.scatter(temporario['comprimento_sépala'], temporario['largura_sépala'], color = cores[especie])

# Titulo e etiquetas e legenda
ax.set_xlabel('Comprimento da sépala',fontsize= 14)
ax.set_ylabel('largura da sépala',fontsize= 14)
ax.set_title('Comprimento da sépala x Largura Sépala', fontsize=20, pad = 15)
ax.legend(cores, loc ='best', fontsize = 15)

# Alterando o tamanho dos ticks dos labels
ax.tick_params(labelsize=15)

```

### Adicionando marcadores 

```python
# Criando a região do grafico
fig = plt.figure(figsize = (12,6))
ax = fig.add_axes([0, 0, 1, 1])

##### Grafico de dispersão #####

# Cores : diferenciando as especias por cores
cores = {'Iris-setosa':'darkorange', 'Iris-versicolor':'r','Iris-virginica':'b'}

for especie in dados['espécie'].unique():
    temporario = dados[dados['espécie'] == especie]
    ax.scatter(temporario['comprimento_sépala'], temporario['largura_sépala'],
               color = cores[especie], marker=marcadores[especie], s= 100)

# Titulo e etiquetas e legenda
ax.set_xlabel('Comprimento da sépala',fontsize= 14)
ax.set_ylabel('largura da sépala',fontsize= 14)
ax.set_title('Comprimento da sépala x Largura Sépala', fontsize=20, pad = 15)
ax.legend(cores, loc ='best', fontsize = 15)

# Alterando o tamanho dos ticks dos labels
ax.tick_params(labelsize=15)

# Marcadores: s = aumenta o tamanho do marcador
marcadores = {'Iris-setosa':'x', 'Iris-versicolor':'o','Iris-virginica':'v'}
```

### Box-plot

```python
# Criando a região do grafico
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0, 0, 1, 1])

##### Box-plot #####
ax.boxplot(dados['largura_pétala'])

# Titulo e etiquetas e legenda
ax.set_title('Box plot', fontsize=20, pad = 15)
ax.set_xticklabels(['largura_pétala'], fontsize= 15)
```

#### Criando um box-plot para cada variável

```python
# Criando a região do grafico
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0, 0, 1, 1])


##### Box-plot #####
ax.boxplot(dados.drop('espécie', axis = 1)) 
# dependendo, temos que adicionar .value() pois ele pode plotar um box-plot para cada exemplo
# axis = 1, coluna
# com o drop tiramos o especies porque a especie nao tem uma distribuicao, ela e categorica e nao numerica


# Titulo e etiquetas e legenda
ax.set_title('Box plot', fontsize=20, pad = 15)
ax.set_xticklabels(dados.drop('espécie',axis =1).columns, fontsize= 15)
```

#### Colocando cores

```python
# Criando a região do grafico
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0, 0, 1, 1])


##### Box-plot #####
caixas = ax.boxplot(dados.drop('espécie', axis = 1), patch_artist=True) 
# dependendo, temos que adicionar .value() pois ele pode plotar um box-plot para cada exemplo
# axis = 1, coluna
# com o drop tiramos o especies porque a especie nao tem uma distribuicao, ela e categorica e nao numerica


# Titulo e etiquetas e legenda
ax.set_title('Box plot', fontsize=20, pad = 15)
ax.set_xticklabels(dados.drop('espécie',axis =1).columns, fontsize= 15)

# Cores : patch_artist coloca cores dentro do box-plot
cores = ['darkorange', 'r','b']


for caixas, cor in zip(caixas['boxes'], cores): 
    # boxes = o box-plot vai retornar dentro dessa varialel as caixas
    caixas.set(color = cor)
    
```

#### No problema, foi soliciatado que os outliers tivessem um marcador x

```python
# Criando a região do grafico
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0, 0, 1, 1])


##### Box-plot #####
caixas = ax.boxplot(dados.drop('espécie', axis = 1), patch_artist=True) 
# dependendo, temos que adicionar .value() pois ele pode plotar um box-plot para cada exemplo
# axis = 1, coluna
# com o drop tiramos o especies porque a especie nao tem uma distribuicao, ela e categorica e nao numerica


# Titulo e etiquetas e legenda
ax.set_title('Box plot', fontsize=20, pad = 15)
ax.set_xticklabels(dados.drop('espécie',axis =1).columns, fontsize= 15)


# Cores : patch_artist coloca cores dentro do box-plot
cores = ['darkorange', 'r','b']
  
for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color=cor)
# o box plot retorna para o boxes a cor



# marcardores
for outlier in caixas['fliers']:
    outlier.set(marker='x', markersize=8)
    
# grid
ax.grid(True)
    
```

### Histograma

```python
# Criando a região do grafico
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0, 0, 1, 1])


##### Histograma #####
ax.hist(dados['comprimento_pétala'])

# Titulo e etiquetas e legenda
ax.set_xlabel('Comprimento das pétalas', fontsize = 15)
ax.set_title('Histograma', fontsize = 20, pad = 10)
ax.set_ylabel('Frequência', fontsize = 15)
ax.grid(True)
```

#### Aumentando o número de classes para o histograma (bins)

```python
# Criando a região do grafico
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0, 0, 1, 1])


##### Histograma #####
ax.hist(dados['comprimento_pétala'], bins = 22)

# Titulo e etiquetas e legenda
ax.set_xlabel('Comprimento das pétalas', fontsize = 15)
ax.set_title('Histograma', fontsize = 20, pad = 10)
ax.set_ylabel('Frequência', fontsize = 15)
ax.grid(True)
```

#### Mudando o sentido do histograma (orientation)

```python
# Criando a região do grafico
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0, 0, 1, 1])


##### Histograma #####
ax.hist(dados['comprimento_pétala'], bins = 22, orientation='vertical') # na vertical fica melhor

# Titulo e etiquetas e legenda
ax.set_xlabel('Comprimento das pétalas', fontsize = 15)
ax.set_title('Histograma', fontsize = 20, pad = 10)
ax.set_ylabel('Frequência', fontsize = 15)
ax.grid(True)
```

#### Trocando frequencia por densidade

```python
# Criando a região do grafico
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0, 0, 1, 1])


##### Histograma #####
ax.hist(dados['comprimento_pétala'],
        bins = 22, orientation='vertical',
        density= True) 

# Titulo e etiquetas e legenda
ax.set_xlabel('Comprimento das pétalas', fontsize = 15)
ax.set_title('Histograma', fontsize = 20, pad = 10)
ax.set_ylabel('Densidade', fontsize = 15)
ax.grid(True)
```

### Vamos obter a média e o desvio padrão

```python
# Criando a região do grafico
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0, 0, 1, 1])


##### Histograma #####
ax.hist(dados['comprimento_pétala'],bins = 22, orientation='vertical') 

# Titulo e etiquetas e legenda
ax.set_xlabel('Comprimento das pétalas', fontsize = 15)
ax.set_title('Histograma', fontsize = 20, pad = 10)
ax.set_ylabel('Frequência', fontsize = 15)
ax.grid(True)

# Media e desvio padrao : devemos ter frequencia e nao densidade
mu = dados['comprimento_pétala'].mean()
sigma = dados['comprimento_pétala'].std()
ax.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),xy=(6.2, 21), fontsize=20)
```

### Traçar uma reta limitnado a média e a mediana

```python
# Criando a região do grafico
fig = plt.figure(figsize=(18,10))
ax = fig.add_axes([0, 0, 1, 1])


##### Histograma #####
ax.hist(dados['comprimento_pétala'],bins = 22, orientation='vertical') 

# Titulo e etiquetas e legenda
ax.set_xlabel('Comprimento das pétalas', fontsize = 15)
ax.set_title('Histograma', fontsize = 20, pad = 10)
ax.set_ylabel('Frequência', fontsize = 15)
ax.grid(True)


# Media e desvio padrao : devemos ter frequencia e nao densidade
# xy = posicao do texto
mu = dados['comprimento_pétala'].mean()
sigma = dados['comprimento_pétala'].std()
ax.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),xy=(6.2, 21), fontsize=20)
# barra invertida antes do mu e sigma deixa o simbolo grego


# limitadores

# media
ax.axvline(mu, color = 'k', linestyle = '--')
ax.annotate('Média', xy = (mu - 0.4, 28), fontsize=20)

#mediana
ax.axvline(dados['comprimento_pétala'].median(), color = 'g', linestyle = '--')
ax.annotate('Mediana', xy = ( dados['comprimento_pétala'].median(), 28), fontsize=20)
```

### Comprimento das pétalas das iris-versicolor

```python
# Criando a região do grafico
fig = plt.figure(figsize=(18,10))
ax = fig.add_axes([0, 0, 1, 1])

# Separação das especies
iris_versicolor = dados[dados['espécie'] == 'Iris-versicolor']

 
##### Histograma #####
ax.hist(iris_versicolor['comprimento_pétala'], bins = 22, orientation='vertical') 

# Titulo e etiquetas e legenda
ax.set_xlabel('Comprimento das pétalas da espécie iris-versicolor', fontsize = 20)
ax.set_title('Histograma da iris-versicolor', fontsize = 25, pad = 10)
ax.set_ylabel('Frequência', fontsize = 20)
ax.grid(True)


# Media e desvio padrao : devemos ter frequencia e nao densidade
# xy = posicao do texto
mu = iris_versicolor['comprimento_pétala'].mean()
sigma = iris_versicolor['comprimento_pétala'].std()
ax.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),xy=(4.8, 6), fontsize=25)
# barra invertida antes do mu e sigma deixa o simbolo grego


# limitadores

# media
ax.axvline(mu, color = 'b', linestyle = '--')
ax.annotate('Média', xy = (mu - 0.14, 7.1), fontsize=25, color ='b')

#mediana
ax.axvline(iris_versicolor['comprimento_pétala'].median(), color = 'r', linestyle = '--')
ax.annotate('Mediana', xy = ( iris_versicolor['comprimento_pétala'].median(), 7.1), fontsize=25, color='r')


# Salvar a imagem
# o bbox retira os espacos em branco na imagem, "tight" deixa junto a imagem
fig.savefig('histograma_iris_versicolor.png', bbox_inches = 'tight')

```
