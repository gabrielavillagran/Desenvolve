<p style='font-size: 22px; line-height: 2; margin: 10px 50px; text-align: justify; color:purple;'> <b> 3.2 Conhecendo a função Agilizando com re.compile</b>
    
<a href = "https://docs.python.org/3/library/re.html" style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'> Documentação</a>
    
    
<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Nós estávamos passando para uma função uma string direto, mas esse não é o único jeito de criar uma Regex. Nós podemos computar antes e utilizar essa função para chamar outras.

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Tem uma função dentro da biblioteca chamada re.search. Ela lê uma string, transforma em uma Regex. Eu passo para ela a string 70, porque vou jogar uma série de números terminando em 70. A search vai transformar essa string numa Regex e buscar o padrão que passamos, que seria nosso texto. Ela vai caminhar em cada caractere e retornar a posição que encontrou.

```python
# retorna a posição do numero 70
re.search(r"70", '389429047422984274489471231228923424242421370')
```

```python
# Função que transforma a string em uma regex direto
regex = re.compile(r'70')

# Chamando função search
regex.search('111111111111111111111111111111111111111170')
```

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'> Descobrindo a diferença entre eles já que retornou o mesmo resultado

```python
# Comparando o tempo de processamento dos dois métodos
from timeit import timeit

setup = """import re """
timeit("""re.search(r"70", '111111111111111111111111111111111111111170')""", setup)
```

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'> O tempo de execução desse método foi de 0.713s, não apenas em uma execução pois seria muito lento, mas em um milhão de execuções, agora vamos comparar com o outro método.

```python
setup = """import re 
regex = re.compile(r'70')"""

timeit("""regex.search('389429047422984274489471231228923424242421370')""", setup)
```

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>O tempo de execução é de 0.205s, é muito menor o tempo, para um texto pequeno não tem muita diferença mas para uma base de dados grande deve ser considerado o tempo. Então fica como boa prática utilizar a função <i><b>re.compile</b></i> para criar a regex e depois chamar outras funções através da variáveis do re.compile.
    
<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Só para deixar claro, o que acontece nesse método é que a cada chamada da função ele pega a string e transforma numa Regex. No outro ele faz isso uma única vez. Essa é a diferença.

<p style='font-size: 22px; line-height: 2; margin: 10px 50px; text-align: justify; color:purple;'> <b> 3.3 Regex: Removendo tag HTML</b>

```python
def remover(textos, regex):
    if type(textos) == str:
        # 1º parâmetro é o que será substitutido na tag HTML de um único texto
        # 2ª parâmetro é o local onde será realizada a substituição
        return regex.sub('', textos)
    else:
        # Percorrer a lista contendo todos os textos e remover a tag html de cada texto
        return [regex.sub('', texto) for texto in textos]
```

```python
# Removendo tags das questões em português 
regex_html = re.compile(r'<.*?>')
questao_sem_tag = remover(questao_portugues, regex_html)
print(questao_sem_tag)
```

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Foram removidos os códigos, mas precisavamos que fosse removido também os trechos com códigos, mas se usarmos as tags html para a detecção desses códigos será mais simples de construir a regex.

```python
# Removendo tags das questões em português 
questao_ingles[5]
```

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Repare que quando temos a introdução do código, temos uma <b>tag code</b>. Quando temos o encerramento, temos a <b>tag /code</b>. Podemos criar uma Regex que identifica esse padrão e ou remove tudo que está lá dentro ou substitui por outra coisa. Para isso, vamos utilizar as tags html. Vamos remover o código e depois a tag html.

```python
def substituir_codigo(textos, regex):
    if type(textos) == str:
        # 1º parâmetro é o que será substitutido na tag HTML de um único texto
        # Foi substituido pela palavra CODE
        # 2ª parâmetro é o local onde será realizada a substituição
        return regex.sub('CODE', textos)
    else:
        # Percorrer a lista contendo todos os textos e remover a tag html de cada texto
        return [regex.sub('CODE', texto) for texto in textos]
```

```python
# Criação da regex que retira o code e a quebra de linha, criaremos um grupo para retirar a quebra
# foi criado um subgrupo para \n para ele interpretar como uma coisa só, por isso entre ()
regex_codigo = re.compile(r'<code>(.|(\n))*?</code>')


# Questão sem código, passando a função criada
questao_sem_code = substituir_codigo(questao_portugues[5], regex_codigo)
print(questao_sem_code)
```

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'> Como podemos ver, agora a questão não possui os códigos, onde tinham códigos foram substituidos pela palavra CODE.
    
<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Agora, o que podemos fazer? Precisamos aplicar isso para todos os nossos dados. Vamos pegar os dados em português, por exemplo, e criar mais uma coluna sem a informação de tags e código. Isso vai ser minha função substituir código recebendo como parâmetro todos os dados em português e a Regex código.

<p style='font-size: 22px; line-height: 2; margin: 10px 50px; text-align: justify; color:purple;'> <b> 3.4 Remoção de dígitos e pontuações</b>
    
<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Vamos remover as pontuações. Para isso, iremos utilizar um outro metacaractere interessante das expressões regulares. É o <b><i>/w</i></b>. Ele combina com qualquer metacaractere alfanumérico. Isso vai ser interessante porque se de alguma maneira conseguirmos pegar o complementar disso, tudo que não for alfanumérico removemos.

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Vamos criar a Regex e utilizar a função remover para remover as pontuações. Eu vou pegar a Regex, que vai ser igual compile, e depois passo a criar mina Regex, que se vocês lembram da ideia de conjunto de caracteres, ele tem uma característica bem bacana. Se eu passo o acento circunflexo dentro dos meus colchetes, ele faz com que o que seja complementar seja complementar da minha Regex que está ali dentro.

```python
# o ^ faz com que o que for combinado com ele seja complementar do caracter dentro dos []
# ela combina com todos os caracteres menos os caracteres não alphanumericos.
regex_portuacao = re.compile(r'[^ \w]')
print(remover(questoes_ingles_sem_code_tag[0], regex_portuacao))
```

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'> O que aconteceu ? Estou pegando o complementar, só que também estou tirando os espaços, a quebra de linha, que virou um bolo só e no fim vai virar uma bagunça. O que posso fazer é utilizar outro conjunto aqui de caracteres que é o /s, que combina com todos os espaços.

```python
# Sera utilizado também o  \s
regex_portuacao = re.compile(r'[^ \w\s]')
print(remover(questoes_ingles_sem_code_tag[0], regex_portuacao))
```

```python
# Sera utilizado também o  \s
# Testando em espanhol para ver as pontuações, foram retiradas
regex_pontuacao = re.compile(r"[^\w\s]")
print(remover(questoes_espanhol_sem_code_tag, regex_pontuacao))
```

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'> Observa-se também que tem palavras que estão diferenciadas pois estão em <i>maiúsculas em minusculas</i>, como a palavras <u>Las e las</u>.

```python
def minusculo(textos):
    if type(textos) == str:
        return textos.lower()
    else:
        return [texto.lower() for texto in textos]
print(minusculo(questoes_espanhol_sem_code_tag))
```

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'> Agora todas as frases foram transformadas em minúsculas. Será criada agora uma regex que retire os dígitos.

```python
# Criando a regex de retirar dígitos
regex_digitos = re.compile(r'\d+')

# Os dígitos foram retirados
print(remover('Teste \n 121232 Textoo 9893', regex_digitos))
```

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'> Enquanto realizamos a remoção do dígito, ficamos com dois espaços. Se eu tivesse uma quebra de linha, teria Teste, a quebra de linha e um espaço depois que não é interessante. Posso criar uma Regex para remover isso. É bem simples.

```python
# Criando a regex de retirar espaços
regex_espaco = re.compile(r" +")

# Retirando a quebra de linha
regex_quebra_linha = re.compile(r"(\n)")

# Criando função se substituir espaços
def substituir_por_espaco(textos, regex):
    if type(textos) == str:
        return regex.sub(" ", textos)
    else:
        return [regex.sub(" ", texto) for texto in textos]

# Os grandes espaços foram substitudos por apenas 1 espaço
print(substituir_por_espaco('Teste \n 121232 Texto \n        9893', regex_espaco))
print('')
# As quebras de linha foram retiradas
print(substituir_por_espaco('Teste \n 121232 Texto \n        9893', regex_quebra_linha))
```

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'> Aplicando todos os tratamentos nos dados em português

```python
# Aplicando para os dados em portugues, retirando as pontuações
questoes_port_sem_pont = remover(dados_portugues.sem_code_sem_tag, 
                                 regex_pontuacao)
# Colocando em minúsculo
questoes_port_sem_pont_minus = minusculo(questoes_port_sem_pont)
questoes_port_sem_pont_minus_dig = remover(questoes_port_sem_pont_minus,
                                          regex_digitos)

# Retirando as quebras de linha
questoes_port_sem_quebra_linha = substituir_por_espaco(questoes_port_sem_pont_minus_dig,
                                                       regex_quebra_linha)
# Retirando os espaços duplicados
questoes_port_sem_espaco_duplicado = substituir_por_espaco(questoes_port_sem_quebra_linha,
                                                          regex_espaco)

dados_portugues["questoes_tratadas"] = questoes_port_sem_espaco_duplicado
```

<p style='font-size: 22px; line-height: 2; margin: 10px 50px; text-align: justify; color:purple;'> <b> 3.5 NLTK e Bigrams</b>
    
<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'><a href = "https://www.nltk.org/_modules/nltk/model/ngram.html" style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'> Documentação Ngrams</a>

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify;'>Ê uma forma de manter uma certa sequência no nosso texto.<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify;'>Esse conceito irá reforçar no nosso texto algumas palavras importantes, além de manter uma estrutura sequencial (seja ela de duas ou três palavras). Porém, manter isso tem um custo: se temos uma frase vetorizada com 50 dimensões, por exemplo, vamos dobrar essas dimensões utilizando o Ngrams, chegando em 100 dimensões. Portanto, devemos tomar cuidado e analisar se esse processo realmente vale a pena.

```python
! pip install nltk
```

```python
from nltk.util import bigrams

texto_teste = "Gabriela"

# Printando o gerador do bigrama
bigrams(texto_teste)

# Passando o objeto para uma lista para vermos a formaçáo dos bigramas
print(list(bigrams(texto_teste)))
```

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify;'>Agora que temos os bigramas, verificamos que a letra inicial e a final só se repetem uma vez. Isso gera um problema na hora de calcular as probabilidades. Uma forma de solucionar esse problema é através dos caracteres falsos.Adicionando o <b><i>fake char</i></b> todas as letras das minhas palavras vão aparecer duas vezes na construção dos bigramas. Para isso, podemos utilizar um pacote chamado language model, do NLTK. É bem interessante, vale a pena dar uma lida. Vamos importar o método.

```python
from nltk.lm.preprocessing import pad_both_ends

# Passando a funçáo pad_both_ends e n =2 pois é um bigrama
print(list(bigrams(pad_both_ends(texto_teste, n = 2))))
```

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify;'>Agora temos os bigramas com o fake chars adicionados, arrumamos uma maneira do nosso modelo de linguagem identificar quando ê inicio e final de uma palavra, gerando por consequência a distribuição de probabilidades correta.

A partir desse cálculo, podemos pensar numa forma de calcular a perplexidade de uma frase inteira. O próximo passo é criar as funções que vão treinar nosso modelo e depois criar função de teste de modelo através da perplexidade.

<p style='font-size: 22px; line-height: 2; margin: 5px 50px; text-align: justify; color:purple;'> <b> 3.10 Função para treinar ML's</b>

```python
def treinar_modelo_mle(lista_textos):
    todas_questoes = " ".join(lista_textos)
    todas_palavras = WhitespaceTokenizer().tokenize(todas_questoes)
    
    # Criando os bigramas e adicionando os fake chars 
    bigrams, vocabulario = padded_everygram_pipeline(2, todas_palavras)
    
    # Criação do modelo
    modelo = MLE(2)
    
    # Treinando o modelo
    modelo.fit(bigrams, vocabulario)
    
    # retornando o modelo
    return modelo
```

```python
# Passando a função
modelo_port_2 = treinar_modelo_mle(X_train_port)

# Calculando a perplexidade
# Testando a perplexidade do texto boa tarde
print(modelo_port_2.perplexity(palavras_bigramns[0]))

print(modelo_port_2.perplexity(palavras_bigramns[1]))

# A função deu certo, deu igual ao anterior, então será usado para os modelos em ingles e espanhol
```

```python
# Passando a função para os dados em ingles
modelo_ing_2 = treinar_modelo_mle(X_train_ing)

# Testando a perplexidade do texto boa tarde
print(modelo_ing_2.perplexity(palavras_bigramns[0]))

print(modelo_ing_2.perplexity(palavras_bigramns[1]))

# Se saiu melhor pois a palavra é em inglês
```

```python
# Passando a função para os dados em espanhol
modelo_esp_2 = treinar_modelo_mle(X_train_esp)

# Testando a perplexidade do texto boa tarde
print(modelo_esp_2.perplexity(palavras_bigramns[0]))

print(modelo_esp_2.perplexity(palavras_bigramns[1]))
```

```python
# Criando função para perplexidade
def calcular_perplexidade(modelo, texto):
    # Contador
    perplexidade = 0
    
    # Separando em palavras
    palavras = WhitespaceTokenizer().tokenize(texto)
    
    # Palavras com o fake char
    palavras_fake_char = [list(pad_both_ends(palavra, n = 2)) for palavra in palavras]

    # Criando os bigramas 
    palavras_bigramns = [list(bigrams(palavra)) for palavra in palavras_fake_char]

    # Acumulando a perplexidade das frases 
    for palavra in palavras_bigramns:
        perplexidade += modelo.perplexity(palavra)
    
    return perplexidade
```

<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify;'> ex: Perplexidade de (x) = 1 / P(x), sendo <b>x</b> uma "palavra" que o modelo não conhece, essa relação da probabilidade e x ocorrer é de 1/0, tende ao infinito.

<p style='font-size: 22px; line-height: 2; margin: 5px 50px; text-align: justify; color:purple;'> <b> 3.11 Modelo de Laplace</b>
    
<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify;'>É uma técnica que faz com que a probabilidade nunca seja zero. Podemos utilizar esse modelo de Laplace, que não deixa a probabilidade ser igual a zero, evitando a perplexidade infinita.

```python
# Importando Laplace
from nltk.lm import Laplace

def treinar_modelo_Laplace(lista_textos):
    todas_questoes = " ".join(lista_textos)
    todas_palavras = WhitespaceTokenizer().tokenize(todas_questoes)
    
    # Criando os bigramas e adicionando os fake chars 
    bigrams, vocabulario = padded_everygram_pipeline(2, todas_palavras)
    
    # Criação do modelo
    modelo = Laplace(2)
    
    # Treinando o modelo
    modelo.fit(bigrams, vocabulario)
    
    # retornando o modelo
    return modelo
```

```python
modelo_ing_Laplace = treinar_modelo_Laplace(X_test_ing)

print(calcular_perplexidade(modelo_ing_Laplace, X_test_port.iloc[0]))

# Agora deu um valor, diferente de inf 
```

```python
modelo_port_Laplace = treinar_modelo_Laplace(X_test_port)

print(calcular_perplexidade(modelo_port_Laplace, X_test_port.iloc[0]))

# Como a frase é em português, o valor deu menor
```

```python
modelo_esp_Laplace = treinar_modelo_Laplace(X_test_esp)

print(calcular_perplexidade(modelo_esp_Laplace, X_test_port.iloc[0]))
```

<p style='font-size: 22px; line-height: 2; margin: 5px 50px; text-align: justify; color:purple;'> <b> 3.12 Atribuir Idiomas</b>
    
<p style='font-size: 16px; line-height: 2; margin: 10px 50px; text-align: justify;'>Vamos criar nossa função que vai receber um conjunto de textos, que são as questões, e precisa atribuir um idioma a eles. Para isso, vamos criar a função atribui idioma. Ela vai receber dois parâmetros. Primeiro o modelo, em que vai ser calculado a perplexidade, e a lista de textos.

```python
def atribui_idioma(lista_textos):
    idioma = []
    for texto in lista_textos:
        portugues = calcular_perplexidade(modelo_port_Laplace, texto)
        ingles = calcular_perplexidade(modelo_ing_Laplace, texto)
        espanhol = calcular_perplexidade(modelo_esp_Laplace, texto)
        
        # Analisar a perplexidade
        if ingles >= portugues <= espanhol:
            idioma.append("portugues")
        elif portugues > ingles < espanhol:
            idioma.append("ingles")
        else:
            idioma.append("espanhol")
    return idioma
```

```python
resultados_espanhol = atribui_idioma(X_test_esp)

# Taxa de acerto
taxa_espanhol = resultados_espanhol.count("espanhol") / len(X_test_esp)
taxa_espanhol
```
