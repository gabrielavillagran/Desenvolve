```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(15, 8)) # configuração do gráfico
```

```python
dados = pd.read_csv('./aluguel_residencial_sem_outliers.csv', sep=';')
dados.head()
```

```python
area = plt.figure()
```

```python
# dentro dessa figura terão 4 gráficos 2 linhas e 2 colunas e ele vai ficar na posição 1
g1 = area.add_subplot(2,2,1)
g2 = area.add_subplot(2,2,2)
g3 = area.add_subplot(2,2,3)
g4 = area.add_subplot(2,2,4)
```

```python
#g1.scatter(dados.Valor, dados.Area) ou como esta abaixo
g1.scatter(dados.Valor, dados.Area)
g1.set_title('Valor X Área')

# fazer uma amostra aleatória (grafico linha) e precisamos arrumar o index
g2.hist(dados.Valor)
g2.set_title('Histograma')

# Grafico de barra
dados_g3 = dados.Valor.sample(100) 
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

#g1.scatter(dados.Valor, dados.Area) ou como esta abaixo
grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor Médio por Tipo')
```
