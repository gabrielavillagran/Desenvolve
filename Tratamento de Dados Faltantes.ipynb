## Tratamento de Dados Faltantes (continuação)

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# imóveis do tipo apartamento que contém condomínio nulo

selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
# justamente o que não queremos, um apartamento sem condomínio

A = dados.shape[0]
dados = dados[~selecao] # inverte o booleano, o que é True vira False vice-versa

B = dados.shape[0]

A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# adiciona 0 no NaN 
# inplace para mudar no DataFrame
# dados.fillna(0, inplace=True)

dados = dados.fillna({'Condomínio':0, 'IPTU':0})
```
