### Fazendo as primeiras visualizações  

```python
import pandas as pd
import matplotlib.pyplot as plt

# faz com que os graficos aparecam de forma automatica
%matplotlib inline

# Aumenta o tamanho das celulas
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))
```

```python
dados = pd.read_csv('./curso-matplotlib-preparando_ambiente/./monitoramento_tempo.csv')
dados.head()
```

### Verificação de valores nulos

```python
dados.info()
```

### Transformacao da coluna data para datetime

```python

dados['data'] = pd.to_datetime(dados['data'])
```

### Gráfico simples

```python
# Grafico para data no eixo x e temperatura para eixo y
plt.plot(dados['data'], dados['temperatura'])
# A temperatura esta em kelvin 
```

### Gráfico ajustando o tamanho da janela

```python
plt.figure(figsize = (20,8))
plt.plot(dados['data'],dados['temperatura'])
```

### Gráfico com título 

```python
# Colocando um titulo
plt.figure(figsize = (12,6))
plt.plot(dados['data'],dados['temperatura'])
plt.title('Temperatura no tempo', fontsize = 16)
```

### Gráfico com fonte maior

```python
fig = plt.figure(figsize = (12,6))
ax = fig.add_axes([0, 0, 1 , 1])
ax.plot(dados['data'], dados['temperatura'])
ax.set_title(' Temperatura no tempo',fontsize = 22)
```

### Gráfico com legendas

```python
fig = plt.figure(figsize = (12,6))
ax = fig.add_axes([0, 0, 1 , 1])
ax.plot(dados['data'], dados['temperatura']) # primeiro x, segundo y
ax.set_title(' Temperatura no tempo',fontsize = 22)
ax.set_xlabel('Anos', fontsize = 16)
ax.set_ylabel('Temperatura', fontsize = 16)
ax.legend(['temperatura'], loc = 'lower right', fontsize =12) # legenda, "lower right" é a posicao da legenda
```

### Vamos adicionar uma seta para visualizarmos melhor o ponto de máximo e mínimo

#### Adicionando uma seta para o máximo

```python
fig = plt.figure(figsize = (12,6))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(dados['data'], dados['temperatura'], color = 'goldenrod')
ax.set_title('Temperatura no tempo', fontsize=20)
ax.set_xlabel('Anos', fontsize=16)
ax.set_ylabel('Temperatura', fontsize=16)
ax.legend(['temperatura'], loc = 'best', fontsize=14)
ax.grid(True) #grid

# limitadores
ax.axhline(max(dados['temperatura']), color = 'k', linestyle = '--')
ax.axhline(min(dados['temperatura']), color = 'k', linestyle = '--')

x1 = dados['data'][dados['temperatura'].idxmax()]
y1 = max(dados['temperatura'])

# vamos deslocar o nome "maximo" mais para esquerda, pois esta muito colado
x2 = dados['data'][dados['temperatura'].idxmax() - 7000] # 7000 horas antes 
y2 = max(dados['temperatura'] - 5)

# texto, arrowprops adiciona uma seta
ax.annotate('Maximo', xy = (x1,y1), fontsize=20, xytext = (x2,y2), arrowprops=dict(facecolor='k'))
```

### Vamos analisar a temperatura de acordo com o dia da semana

```python
temperatura_por_dia_da_semana = dados.groupby('dia_da_semana')['temperatura'].mean()
temperatura_por_dia_da_semana
# As temperaturas parecem muito proximas

# Vamos ordenar
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta','Sábado', 'Domingo']

# Vamos atribuir essa ordem
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana

##### Grafico de barras para termos a temperatura por dia da semana #####

# Criando a região do grafico
fig = plt.figure(figsize = (5,4))
ax = fig.add_axes([0, 0, 1, 1])

# Colocando os dados no eixo
index = nome_dias
ax.bar(index, temperatura_por_dia_da_semana)
ax.set_title('Temperatura por dia da semana', fontsize = '14', pad = 15)
ax.set_xlabel('Dias da semana', fontsize= 12)
ax.set_ylabel('Temperatura Média', fontsize= 12)
```
