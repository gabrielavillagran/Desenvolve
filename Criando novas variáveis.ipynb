## Criando novas variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('./aluguel.csv/./aluguel_residencial.csv', sep=';')
```

```python
dados.head(10)
```

```python
# Obter o Valor bruto
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
# Renomeei o m2 para m²

dados.columns = ['Tipo', 'Bairro','Quartos','Vagas','Suites','Area','Valor','Condominio','IPTU','Valor Bruto','Valor m²']
```

```python
dados
```

```python
# Arrendado o valor por metro quadrado
dados['Valor m²'] = dados['Valor m²'].round(2)
```

```python
dados
```

```python
dados['Valor Bruto m²'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados
```

```python
# Criar uma lista com a idenficação das casas

casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

## Excluindo variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m²', 'Valor Bruto', 'Valor Bruto m²']])
```
