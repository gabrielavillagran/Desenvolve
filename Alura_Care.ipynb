<p style='font-size: 22px; line-height: 2; margin: 10px 50px; text-align: justify; color:red;'><b>  3. Criando os *Datasets* de Treino e Teste </b></p>

<a style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;' href = "https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html"> Documentação </a>

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'><b>Dados de Treino:</b>
Conforme podemos imaginar, dados de treino são os dados que serão apresentados ao algoritmo de machine learning para criação do modelo. Estes dados costumam representar cerca de 70% da totalidade dos dados. Como o Dataset é pequeno, usarei 80% para treino.</p><br>
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'><b>Dados de Teste:</b>
Serão apresentados ao modelo após a sua criação, simulando previsões reais que o modelo realizará, permitindo assim que o desempenho real seja verificado. Estes dados costumam representar cerca de 20% da totalidade dos dados.Como o Dataset é pequeno, usarei 10% para teste.</p><br>

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'>Referência : https://didatica.tech/dados-de-treino-e-teste/</p>

```python
from sklearn.model_selection import train_test_split
from numpy import random

SEED= 1234
random.seed(SEED)

valores_exames = resultados_exames_v1.drop(columns =['id','diagnostico'], axis = 1)
diagnosticos = resultados_exames_v1['diagnostico']


X_treino, X_teste, y_treino, y_teste = train_test_split(valores_exames, diagnosticos, test_size=0.30)
```

```python
X_treino.head()
```

```python
y_treino.head()
```

<p style='font-size: 22px; line-height: 2; margin: 10px 50px; text-align: justify; color:red;'><b>3.1 Criando modelo de classificação *Random Forest* </b>

<a style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;' href = "https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html"> Documentação </a>

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Criando o modelo
classificador = RandomForestClassifier(n_estimators=100)
classificador
```

```python
# Treinando o modelo
classificador.fit(X_treino,y_treino)
```

```python
print('Resultado da classificação %.2f%%' %(classificador.score(X_teste, y_teste) * 100))
```

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'>Precisamos verificar se 92.40% é um bom número para classificação. Isso porque tal valor será a nossa <b>baseline</b> para quando começarmos a reduzir a dimensionalidade. Para fazermos esse processo sem perder a qualidade, primeiro precisamos saber se a nossa base comparativa é mesmo um bom resultado.

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>O <b>ViolinPlot</b> combina a visualização do boxplot juntamente com um plot de densidade (similar ao que vemos nos histogramas).A vantagem do uso do ViolinPlot é que, além das informações que o boxplot já traz consigo, existe a exibição da densidade dos dados. Na figura, a área formada em torno do boxplot representa a distribuição dos dados. Os locais de maior área (onde é mais “gordo” o gráfico), existe uma grande concentração de dados, enquanto em locais de menor área (onde o gráfico é mais “fino”) existe uma baixa concentração de dados.

<p style='font-size: 22px; line-height: 2; margin: 10px 50px; text-align: justify; color:red;'><b>5.1 Padronizando o gráfico utilizando *StandardScaler*</b>
    
<a style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;' href="https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html">Documentação</a>
    
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Padroniza as features removendo a média e escala a variância a uma unidade.

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Isso significa que para cada feature, a média seria 0, e o Desvio Padrão seria 1. Desta forma, as features são padronizadas, tornando-as mais manejáveis para nossos modelos. Mais uma vez, modelos <b>Não</b> Baseados em Árvores de Decisão se beneficiam mais deste tipo de padronização.
    
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'> O eixo <b>y</b> vai de 0 à <b>2500</b>. Será colocado um limite ou padrão que permita trabalhar com esses dados de maneira mais eficiente.

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Padronizador
padronizador = StandardScaler()
padronizador.fit(valores_exames)
valores_exames_v2 = padronizador.transform(valores_exames)


# É um array, então para passar o iloc depois, ele precisa ser um DataFrame
valores_exames_v2 = pd.DataFrame(data = valores_exames_v2, columns=valores_exames.keys())
```

```python
# Tamanho do plot
plt.figure(figsize=(10,8))


# Juntando diagnosticos com valores_exames
dados_plot = pd.concat([diagnosticos, valores_exames_v2.iloc[:,0:10]], axis = 1)

# Agrupando os exames
dados_plot = pd.melt(dados_plot,
                     id_vars='diagnostico',
                     var_name='exames',
                     value_name='valores')
dados_plot


sns.violinplot(x = "exames",
               y = "valores",
               hue = 'diagnostico',
               split = True,
               data=dados_plot)
plt.xticks(rotation = 90)
```

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'> O <b>split</b>= True plotará  no mesmo lado do violin o diagnóstico de benigno e maligno. No eixo "x" temos os exames, no "y" os valores normalizados, e no corpo do gráfico temos o violin plot. Mas como interpretar essas informações? Como as legendas apontam, do lado esquerdo, em azul, temos os cânceres do tipo maligno, e do lado direito, em laranja, os do tipo benigno. A parte mais alta da curva são os valores que mais ocorrem para cada tipo.

<p style='font-size: 22px; line-height: 2; margin: 10px 50px; text-align: justify; color:red;'><b>5.2 Função para a plotagem</b>

```python
def grafico_violino(valores, inicio, fim):
    # Tamanho do plot
    plt.figure(figsize=(10,8))

    # Juntando diagnosticos com valores_exames
    dados_plot = pd.concat([diagnosticos, valores_exames_v2.iloc[:,inicio:fim]], axis = 1)

    # Agrupando os exames
    dados_plot = pd.melt(dados_plot,
                     id_vars='diagnostico',
                     var_name='exames',
                     value_name='valores')
    dados_plot


    sns.violinplot(x = "exames",
               y = "valores",
               hue = 'diagnostico',
               split = True,
               data=dados_plot)
    plt.xticks(rotation = 90)
```

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Será chamada a função abaixo, que mostrará a  visualização do gráfico de violino dos exames de <b>10 a 21</b>.

<p style='font-size: 22px; line-height: 2; margin: 10px 50px; text-align: justify; color:red;'><b>6. Criando novo modelo de classificação *Random Forest* </b>

```python
valores_exames_v3 = valores_exames_v2.drop(columns=['exame_4', 'exame_29'], axis = 1)
```

```python
def classificar(valores):
    # Criando uma semente
    SEED= 1234
    random.seed(SEED)
    
    X_treino, X_teste, y_treino, y_teste = train_test_split(valores_exames_v3, diagnosticos, test_size=0.30)
    
    # Criando o modelo
    classificador = RandomForestClassifier(n_estimators=100)
    
    # Treinando o modelo
    classificador.fit(X_treino,y_treino)
    
    # Mostrando o resultado
    print('Resultado da classificação %.2f%%' %(classificador.score(X_teste, y_teste) * 100))
    
```

```python
classificar(valores_exames_v3)
```

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'> O resultado não teve impacto negativo em comparação com a <b>Baseline</b>.

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>A multicolinearidade ocorre quando o modelo inclui vários fatores correlacionados não apenas à sua variável de resposta, mas também uns aos outros. Em outras palavras, resulta quando você tem fatores que são, de certa forma, um pouco redundantes. As correlações próximas a 1 serão retiradas para depois estimarmos o modelo.
    
fonte:https://blog.minitab.com/pt/basta-lidando-com-a-multicolinearidade-na-analise-de-regressao#:~:text=A%20multicolinearidade%20ocorre%20quando%20o,certa%20forma%2C%20um%20pouco%20redundantes.

```python
variaveis_correlacionadas = matriz_correlacao_v2[matriz_correlacao_v2 > 1]
variaveis_correlacionadas
```

<p style='font-size: 22px; line-height: 2; margin: 10px 50px; text-align: justify; color:red;'><b>8. Selecionando Features</b>

<p style='font-size: 22px; line-height: 2; margin: 10px 50px; text-align: justify; color:red;'><b>8.1 Selecionando as melhores Features com *SelectKBest*</b>
    
<a style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;' href=" https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.SelectKBest.html">Documentação</a>  
    
<a href="https://medium.com/data-hackers/como-selecionar-as-melhores-features-para-seu-modelo-de-machine-learning-faf74e357913" style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Material complementar</a> 


```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
# Criação do modelo
selecionar_kmelhores = SelectKBest(chi2, k= 5)
```

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Será utilizado <b>5</b> como valor de k. Ou seja, de exames que eram realizados para chegar a um diagnóstico, queremos realizar apenas 5, mantendo a qualidade de predição do algoritmo. Armazenaremos o retorno dessa chamada em uma variável <b>selecionar_kmelhores</b> e exibiremos o seu conteúdo na tela.
    
    
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>A função qui-quadrado não aceita valores negativos, e existem vários deles, por exemplo, na coluna "exame_5". Uma alternativa para corrigirmos isso seria voltarmos para os valores não-normalizados que tínhamos no conjunto valores_exames. Sendo assim, teremos que remover deste conjunto todas as colunas que não estão presentes em valores_exames_v5. Para isso, faremos um drop() das colunas "exame_4", "exame_29", "exame_3" e "exame_24", e armazenaremos o retorno em uma variável valores_exames_v6.

```python
# o exame 4 e 29 são os constantes e o 3 e 24 são os que tem alta correlação

valores_exames_v5 = valores_exames.drop(columns=['exame_4', 'exame_29', 'exame_3', 'exame_24'])
valores_exames_v5
```

```python
# Criando o treino e o teste

SEED= 1234
random.seed(SEED)

X_treino, X_teste, y_treino, y_teste = train_test_split(valores_exames_v5,
                                                        diagnosticos,
                                                        test_size=0.30)

selecionar_kmelhores.fit(X_treino, y_treino)


# Transformação dos dados
# Agora teremos só selecionado 5 features

kbest_treino = selecionar_kmelhores.transform(X_treino)

kbest_teste = selecionar_kmelhores.transform(X_teste)
```

```python
# Conferindo o número de colunas
print(kbest_treino.shape, kbest_teste.shape)
```

```python
classificador = RandomForestClassifier(n_estimators=100, random_state=1234)
classificador.fit(kbest_treino, y_treino)
```

```python
print('Resultado da classificação %.2f%%' %(classificador.score(kbest_teste, y_teste) * 100))
```

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Na realidade da Alura Care, supondo que cada exame tem o mesmo custo, tivemos uma redução de 85%! <p style='font-size: 22px; line-height: 2; margin: 10px 50px; text-align: justify; color:red;'><b>8.1.1 Matriz de Confusão</b></p>
<a style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;' href=" https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html">Documentação</a> 
    

 <p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>A Matriz de Confusão é uma tabela onde facilmente identificamos todos os quatro tipos de classificação do modelo de classificação binário (isto é, com apenas dois valores distintos na variável resposta). Com ela, facilmente podemos calcular valores como acurácia, especificidade, sensibilidade, etc. </p>
    
    
<li style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'><b>Verdadeiros Positivos (TP)</b>são observações cujo valor real é positivo e o valor previsto é positivo, isto é, o modelo acertou</li>

<li style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'><b>Verdadeiros Negativos (TN)</b> são observações cujo valor real é negativo e o valor previsto é negativo, isto é, o modelo acertou</li>

<li style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'><b>Falsos Positivos (FP)</b> são casos em que o resultado correto é negativo entretanto o resultado obtido é positivo, isto é, o modelo errou</li>
<li style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'><b>Falsos Negativos (FN)</b> são casos em que o resultado correto é positivo entretanto o resultado obtido é negativo, isto é, o modelo errou</li>
    
<p style='font-size: 20px; line-height: 2; margin: 10px 50px; text-align: justify; color:Red;'><b>Métricas</b>
    
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Acurácia, Sensibilidade e Especificidade se referem a métricas de avaliação dos modelos, queremos olhá-las para checar quão bem o modelo está fazendo as classificações. Quanto melhor o modelo, mais altos são os valores dessas métricas.

<li style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'><b>Acurácia:</b> é a proporção de acertos do modelo. Para calculá-la, pega-se o número total de observações que o modelo acertou e divide-se pelo número total de observações que o modelo previu.
    
<li style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'><b>Sensibilidade:</b> também chamada de recall, é a proporção de casos positivos que foram identificados corretamente. Para calculá-la, toma-se o número de observações que o modelo classificou como positivos corretamente (verdadeiros positivos) e divide-se pelo número total de observações com rótulo positivo (verdadeiros positivos + falsos negativos).
    
<li style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'><b>Especificidade:</b> é a proporção de casos negativos que foram identificados corretamente. Para calculá-la, toma-se o número de observações que o modelo classificou como negativos corretamente (verdadeiros negativos) e divide-se pelo número total de observações com rótulo negativo (verdadeiros negativos + falsos positivos).
    
<li style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'><b>Valor de previsão positivo/negativo:</b> em um problema de classificação binário, as observações pertencem a duas classes, que podem ter quaisquer nomes (bom ou mau, sim ou não, 0 ou 1). No momento do treinamento, essas categorias são sempre codificadas como 0 ou 1, de modo que nos referimos às observações com rótulo 0 como sendo a classe negativa, e as observações com rótulo 1 como sendo a classe positiva. Portanto, quando o modelo diz que uma observação tem valor 0 na variável-resposta, dizemos que esse é um valor de previsão negativo, e quando o modelo diz que uma observação tem valor 1 na variável-resposta, dizemos que esse é um valor de previsão positivo.</li>   
    
    
<li style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'> os elementos <b>i</b> da matriz serão os valores <b>reais</b> (teste) </li>
<li style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>os elementos <b>j</b> da matriz serão os valores <b>preditos</b> </li>

```python
# Criando uma semente
SEED= 1234
random.seed(SEED)

# Criando treino e teste
X_treino, X_teste, y_treino, y_teste = train_test_split(valores_exames_v5, diagnosticos, test_size=0.30)

# Criando o modelo 
classificador = RandomForestClassifier(n_estimators=100)

# Treinando o modelo
classificador.fit(X_treino, y_treino)

selecionador_rfe = RFE(estimator = classificador, n_features_to_select=5, step= 1)
selecionador_rfe.fit(X_treino, y_treino)


# Transformando os dados de treino
rfe_treino = selecionador_rfe.transform(X_treino)

# Transformar  os dados de teste
rfe_teste = selecionador_rfe.transform(X_teste)

# Treinando o modelo
classificador.fit(rfe_treino, y_treino)

# Mostrando na tela
print('Resultado da classificação %.2f%%' %(classificador.score(rfe_teste, y_teste) * 100))
```

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>RFE Cross Validation. O RFECV divide o nosso banco de dados em blocos e aplica o algoritmo RFE, que acabamos de aprender, em cada um desses blocos, gerando diferentes resultados. Dessa forma, O RFECV não só nos informa quantas features precisamos ter para gerar o melhor resultado possível, como também quais features são essas.

```python
from sklearn.feature_selection import RFECV
```

```python
# Criando uma semente
SEED= 1234
random.seed(SEED)

# Criando treino e teste
X_treino, X_teste, y_treino, y_teste = train_test_split(valores_exames_v5, diagnosticos, test_size=0.30, random_state=SEED)

# Criando o modelo 
classificador = RandomForestClassifier(n_estimators=100)

# Treinando o modelo
classificador.fit(X_treino, y_treino)

selecionador_rfecv = RFECV(estimator = classificador, cv = 5, step= 1, scoring="accuracy")
selecionador_rfecv.fit(X_treino, y_treino)


# Transformando os dados de treino
rfecv_treino = selecionador_rfe.transform(X_treino)

# Transformar  os dados de teste
rfecv_teste = selecionador_rfe.transform(X_teste)

# Treinando o modelo
classificador.fit(rfecv_treino, y_treino)

# Mostrando na tela
print('Resultado da classificação %.2f%%' %(classificador.score(rfecv_teste, y_teste) * 100))
```

<a style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;' href=" https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html">Documentação TSNE</a> 

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>t-SNE um algoritmo de redução de dimensionalidade não linear encontra padrões nos dados com base na similaridade de pontos de dados com características, a similaridade de pontos é calculada como a probabilidade condicional de que um ponto A escolheria o ponto B como seu vizinho.
Em seguida, ele tenta minimizar a diferença entre essas probabilidades condicionais (ou semelhanças) no espaço de dimensão superior e dimensão inferior para uma representação perfeita dos pontos de dados no espaço de dimensão inferior.

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>O TSNE é um pouco mais lento que o PCA, e por isso, a depender do tamanho da sua base de dados, resultará em um custo computacional um pouco alto. Como resultado, teremos uma visualização que separa muito bem os dados no espaço, tornando mais claro quais pontos estão próximos de um tipo de câncer maligno ou benigno. Assim, é bem mais simples determinar uma fronteira entre tais classificações.</p>

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Essas técnicas que realizam uma transformação matemática nos dados antes de processá-los, reduzindo suas dimensões, são chamadas justamente de "redução de dimensionalidade". A diferença entre a redução de dimensionalidade e a seleção de features é que a segunda funciona como um funil que filtra as principais características de uma base de dados original.</p>


<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; color:black;'>Ao se utilizar as técnicas de seleção de feature para gerar uma visualização no plano, é preciso escolher dois dos atributos para representar os dados. Manter apenas dois dos diversos atributos, faz com que um grande número de informação seja perdida, resultado em visualizações que podem induzir à conclusões equivocadas.</p>
