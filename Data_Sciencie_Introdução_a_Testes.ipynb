```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(75243)
temp = (nota_media_dos_filmes_com_pelo_menos_10_votos.sample(frac = 1)) # deixamos as notas de forma aleatória

medias = [temp[0:i].mean() for i in range(1,len(temp))] # usando list comprehension
plt.plot(medias)

# Pelo plot, conseguimos ver que uma amostra muito pequena ou mal escolhida
# nos leva a ter o entendimento errado sobre os dados. 
# Analisando o plot teriamos uma média de 3,43
# se fosse uma amostra de 500 notas achariamos que a média seria de 3,52 aprox.

# Concluimos que, a medida que temos mais votos, a média fica mais estável.
```

```python
# Queremos uma maneira numerica de dizer quais filmes fora dessa amostra estao
# entre 3.4 - 3.5 com nivel de confianca de 95% 
from statsmodels.stats.weightstats import zconfint
# passar pro zconfint o x1, o conjunto de dados

zconfint(nota_media_dos_filmes_com_pelo_menos_10_votos)

# Dado nosso conjunto de dados, considerando essa amostra grande o suficiente 
# para aplicarmos esse teste ( curva normal, quantidade de dados grande)
# teremos um intevalo de 3.41 - 3.45 pros filmes em geral, nao apenas para nossa
# amostra
```

```python
""" Usar o Z-test ou T-test ?
De uma maneira geral, usamos o Z-test quanto temos um conjunto maior de dados
e o T-test um conjunto menor.  
O que é considerado grande ? mais de 30 elementos.
"""
```

```python
# Vamos usar o  T-test para ver, mas dara proximo ao Z-test

from statsmodels.stats.weightstats import DescrStatsW

descr_todos_com_10_votos = DescrStatsW(nota_media_dos_filmes_com_pelo_menos_10_votos)
descr_todos_com_10_votos.tconfint_mean()
```

```python
from statsmodels.stats.weightstats import ztest
# o ztest vai me dizer se a media desse conjunto vai ser igual ou nao ao outro
ztest(notas1['rating'], value = 3.43205)

#  8.977723251039489e-18 é um numero MUITO pequeno, de forma que < = 0,05
# entao descartamos a hipotese de que o valor e igual a 3,43.

# O filme Toy Story tem uma media diferente da media dos outros filmes 
```

```python
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

def calcula_teste(i):
    media = temp[0:i].mean()
    stat, p = ztest(temp[0:i], value = 3.4320503405352603) 
    return (i, media, p)

valores = np.array([calcula_teste(i) for i in range(2, len(temp))])
plt.plot(valores[:,0], valores[:,1])
plt.plot(valores[:,0], valores[:,2])
plt.hlines(y = 0.05, xmin = 2, xmax = len(temp), colors = 'r')
```
