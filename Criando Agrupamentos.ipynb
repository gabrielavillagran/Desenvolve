```python
grupo_bairro['Valor'].describe().round(2).head(50)
```

```python
## Com o aggregate escolhi apenas as estatisticas que me interessavam do Describe
grupo_bairro['Valor'].aggregate(['min','max']).rename(columns = {'min':'Mínimo','max':'Máximo'})
```

```python
# Vamos importar o matplotlib para graficamente ficar mais facil de encontrar algum problema nos dados
# Pois estamos lidando com todos os bairros

%matplotlib inline 
# para nos mostrar um gráfico estático na tela
import matplotlib.pyplot as plt 
plt.rc('figure', figsize=(20,10)) # configurar o tamanho do gráfico
```

## Exercício

```python
precos = pd.DataFrame([['Feira', 'Cebola', 2.5], 
                        ['Mercado', 'Cebola', 1.99], 
                        ['Supermercado', 'Cebola', 1.69], 
                        ['Feira', 'Tomate', 4], 
                        ['Mercado', 'Tomate', 3.29], 
                        ['Supermercado', 'Tomate', 2.99], 
                        ['Feira', 'Batata', 4.2], 
                        ['Mercado', 'Batata', 3.99], 
                        ['Supermercado', 'Batata', 3.69]], 
                        columns = ['Local', 'Produto', 'Preço'])
precos
```

```python
produtos = precos.groupby('Produto')
produtos.describe().round(2)
```

```python
estatisticas = ['mean', 'std', 'min', 'max']
nomes = {'mean': 'Média', 'std': 'Desvio Padrão', 'min': 'Mínimo', 'max': 'Máximo'}
produtos['Preço'].aggregate(estatisticas).rename(columns = nomes).round(2)
```

Já conseguimos observar alguns dados que estão variando muito em relação aos demais
